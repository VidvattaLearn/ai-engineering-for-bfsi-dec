```python
# quick assignment on for loop
# we have a product prices list #prices = [100, 200, 300, 400]
# we have to apply a 10% discount on each price and create a new discounted price list
# print the new discounted prices
# solution:
prices = [100, 200, 300, 400]
discounted_prices = []
for price in prices:
  disc_price = price - (price * 0.1)
  discounted_prices.append(disc_price)
print(discounted_prices)
```

```python
# assignment on while loop
# we are building a simple game where a player has to guess secret number between 1 and 5.
# use a while loop to keep asking the player if the number does not match until the correct number is guessed
secret_number =4
guess = 0
while guess != secret_number:
  guess = int(input("Guess the secret number (between 1 and 5):"))
print("Congrats you guessed the correct number")
```

```python
# break statement
files = ["data1.csv","data2.csv","error.log","data3.csv"]
for file in files:
  if file == "error.log":
    print("Error log found! Halting the process")
    break
  print(f"Processing {file}...")
```

```python
# assignment on break statement
# you guyz are processing a list of orders, but if you encounter an order
# with negative value, you need to stop the processing.
# solution:
value = [1, 2, 3, -1, 4]

for item in value:
  if item < 0:
    print("Negative value found! Halting the process")
    break
  print(f"Processing {item}...")
```

```python
# continue statement
print("Print only even numbers")
for num in range(1,6):
  if num % 2 !=0:
    continue
  print(num)
```

```python
# simple assignment on contnue as well
# we are analyzing a list of temperatures, we then want to skip some invalid readings,
# like negative values.
# use for loop with continue to print only valid temperatures
temp=[12,89,90,78,-90,-34]
for i in temp:
  if i<0:
    continue 
  print(f"Valid Temperature:{i}")
```

```python
# assignment on functions
# write a python function to calculate total price of items in a cart, including tax
# fucntion should take as input list of items and tax rate as well.
# return the total price after tax addition per item
# cart = [100, 200, 300] tax_rate = 0.07
def calculate_total(prices, tax_rate):
  total = sum(prices)
  total_with_tax = total + (total * tax_rate)
  return total_with_tax
calculate_total([100, 200, 300],0.07)
```
