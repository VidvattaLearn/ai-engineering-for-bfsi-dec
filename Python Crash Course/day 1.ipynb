```python
score = input("Enter score to check your grade:")
score = int(score)
grade = ''

if score >= 90:
  grade = 'A'
elif score >= 70:
  grade = 'B'
elif score >= 60:
  grade = 'C'
else:
  grade = 'D'

print(f"For the given score:{score} your grade is {grade}")
```
